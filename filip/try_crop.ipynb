```python
import numpy as np
import matplotlib
from matplotlib import pyplot
import os
from mosaic import *
```

```python
#dirname is the directory that contains subdirectories RGB-PanSharpen/ and roads/
#RGB-PanSharpen/ contains original .tif images, and roads/ contains rasterized road data
#it expects that the .tif files in roads/ have the same global position data as original images, see generating_roads notebook
dirname = os.getcwd() + "/spacenet_sample/AOI_2_Vegas_Train/"
```

```python
#create mosaic of images in dirname
m = Mosaic(dirname)
```

```python
#range of coordinates in the global coordinate system
m.minx, m.maxx, m.miny, m.maxy
```

```python
#crop the square with side length of 400 pixels, with one corner at (-115.25, 36.23) and rotated by 30 degrees
cropped_data1 = m.cropImage(400, x = -115.25, y = 36.23, alpha = 30)
```

```python
#if position and angle are unspecified it chooses random values
cropped_data2 = m.cropImage(400)
```
